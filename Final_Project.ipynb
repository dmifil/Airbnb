```python
threshold = 3500

columns_to_remove = data.columns[data.isnull().sum() > threshold]

data = data.drop(columns=columns_to_remove)
```

```python
numeric_columns_with_missing = data.select_dtypes(include='number').columns[data.select_dtypes(include='number').isnull().sum() > 0]


data[numeric_columns_with_missing] = data[numeric_columns_with_missing].fillna(data[numeric_columns_with_missing].median())
```

```python
data.isnull().sum()
```

```python
columns_to_drop = ['neighborhood_overview', 'host_acceptance_rate', 'neighbourhood']
data = data.drop(columns=columns_to_drop)
data['description'] = data['description'].fillna('No description available')
data['host_location'] = data['host_location'].fillna('No host location data')
data['host_neighbourhood'] = data['host_neighbourhood'].fillna('No host neighboourhood data')
```

```python
data['description'] = data['description'].fillna('No description available')
data['host_location'] = data['host_location'].fillna('No host location data')
data['host_neighbourhood'] = data['host_neighbourhood'].fillna('No host neighboourhood data')
```

```python
data['bathrooms_text'] = data['bathrooms_text'].fillna(data['bathrooms_text'].mode().iloc[0])

data['first_review'] = data['first_review'].fillna('No reviews yet')
```

```python
data = data.dropna(subset=['first_review', 'last_review'])
```

# 1. Distribution of Property Types

```python
plt.figure(figsize=(12, 6))
property_type_counts = data['property_type'].value_counts()
sns.barplot(x=property_type_counts.index, y=property_type_counts.values, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Distribution of Property Types')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.show()
```

# 2. Price Distribution

```python
plt.figure(figsize=(12, 6))
sns.histplot(data['price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

# 3. Availability Over Time

```python
availability_cols = ['availability_30', 'availability_60', 'availability_90', 'availability_365']

plt.figure(figsize=(12, 6))
for col in availability_cols:
    sns.lineplot(x=data['last_scraped'], y=data[col], label=col)

plt.title('Availability Over Time')
plt.xlabel('Last Scraped')
plt.ylabel('Availability')
plt.legend()
plt.show()
```

# 4. Review Scores Distribution

```python
score_cols = ['review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness',
              'review_scores_checkin', 'review_scores_communication', 'review_scores_location',
              'review_scores_value']

plt.figure(figsize=(16, 8))
for col in score_cols:
    sns.kdeplot(data[col], label=col, fill=True)
plt.title('Distribution of Review Scores')
plt.xlabel('Score')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# 5. Instant Booking Percentage

```python
plt.figure(figsize=(6, 6))
instant_bookable_counts = data['instant_bookable'].value_counts()
sns.barplot(x=instant_bookable_counts.index, y=instant_bookable_counts.values, palette='pastel')
plt.title('Instant Booking Availability')
plt.xlabel('Instant Bookable')
plt.ylabel('Count')
plt.show()
```

# 6. Host Listings Count vs. Property Type

```python
plt.figure(figsize=(14, 6))
sns.boxplot(x='property_type', y='host_listings_count', data=data, palette='muted')
plt.xticks(rotation=45, ha='right')
plt.title('Host Listings Count by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Host Listings Count')
plt.show()
```

# 7. Geographical Distribution of Listings

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='longitude', y='latitude', hue='property_type', data=data, palette='Set2', alpha=0.6)
plt.title('Geographical Distribution of Listings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Property Type', bbox_to_anchor=(1, 1))
plt.show()
```

# 8. Number of Reviews vs. Review Scores

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='number_of_reviews', y='review_scores_rating', data=data, hue='property_type', palette='viridis')
plt.title('Number of Reviews vs. Review Scores')
plt.xlabel('Number of Reviews')
plt.ylabel('Review Scores Rating')
plt.legend(title='Property Type', bbox_to_anchor=(1, 1))
plt.show()
```

# 9. Accommodation vs. Price and Number of Reviews

```python
plt.figure(figsize=(14, 6))
sns.scatterplot(x='accommodates', y='price', size='number_of_reviews', data=data, hue='property_type', palette='coolwarm', alpha=0.8)
plt.title('Accommodation vs. Price and Number of Reviews')
plt.xlabel('Accommodates')
plt.ylabel('Price')
plt.legend(title='Property Type', bbox_to_anchor=(1, 1))
plt.show()
```
